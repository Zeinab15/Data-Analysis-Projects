```python
#Identify Columns with Invalid Zeros
columns_with_invalid_zeros = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
```

```python
#Replace Zero Values with NaN
for col in columns_with_invalid_zeros :
  df[col] = df[col].replace(0, np.nan)
```

```python
#Check Missing Values
df.isna().sum()
```

```python
# Impute missing values with median
for col in columns_with_invalid_zeros:
   median = df[col].median()
   df[col] =df[col].fillna(median)
```

```python
# CHECK missin values again
df.isna().sum()
```

Explore the Data (EDA)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# correlation matrix


plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
#Boxplot Analysis

features = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI", "DiabetesPedigreeFunction", "Age"]

plt.figure(figsize=(15, 12))
for i, col in enumerate(features):
    plt.subplot(3, 3, i+1)
    sns.boxplot(x="Outcome", y=col, data=df)
    plt.title(f"{col} vs Outcome")
plt.tight_layout()
plt.show()
```

```python
#Scatterplot between Glucose and BMI
sns.scatterplot(data=df, x="Glucose", y="BMI", hue="Outcome")
plt.title("Glucose vs BMI by Outcome")
plt.show()
```

```python
#Histogram of Glucose
sns.histplot(data=df, x="Glucose", hue="Outcome", bins=30, kde=True)
plt.title("Glucose Distribution by Diabetes Outcome")
plt.show()
```

```python
#Histogram of BMI


sns.histplot(data=df, x="BMI", hue="Outcome", bins=30, kde=True)
plt.title("BMI Distribution by Diabetes Outcome")
plt.show()
```
